```python
# 중경1공장 전용 분석 함수들

def plot_distribution(data, column, title=None, bins=30):
    """수치형 변수의 분포를 히스토그램과 박스플롯으로 시각화"""
    if column not in data.columns:
        print(f"❌ '{column}' 컬럼을 찾을 수 없습니다.")
        return
    
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    # 히스토그램
    data[column].dropna().hist(bins=bins, ax=axes[0], alpha=0.7, edgecolor='black')
    axes[0].set_title(f'{title or column} 분포', fontweight='bold')
    axes[0].set_xlabel(column)
    axes[0].set_ylabel('빈도')
    axes[0].grid(True, alpha=0.3)
    
    # 박스플롯
    data[column].dropna().plot(kind='box', ax=axes[1])
    axes[1].set_title(f'{title or column} 박스플롯', fontweight='bold')
    axes[1].set_ylabel(column)
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # 기본 통계
    print(f"📊 {title or column} 기본 통계:")
    print(data[column].describe().round(2))

def compare_by_quality(data, column, top_n=5):
    """품질별로 특정 변수를 비교 분석"""
    if 'quality' not in data.columns or column not in data.columns:
        print(f"❌ 필요한 컬럼을 찾을 수 없습니다.")
        return
    
    # 상위 N개 품질 유형 선택
    top_qualities = data['quality'].value_counts().head(top_n).index
    filtered_data = data[data['quality'].isin(top_qualities)]
    
    plt.figure(figsize=(12, 6))
    
    # 박스플롯
    sns.boxplot(data=filtered_data, x='quality', y=column)
    plt.title(f'중경1공장: 품질별 {column} 비교 (상위 {top_n}개 품질)', fontweight='bold')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # 품질별 통계
    print(f"\n📊 품질별 {column} 통계:")
    quality_stats = filtered_data.groupby('quality')[column].agg(['count', 'mean', 'std', 'min', 'max']).round(2)
    print(quality_stats)

def correlation_analysis(data, columns=None):
    """주요 변수들 간의 상관관계 분석"""
    if columns is None:
        columns = ['p_thick_mm', 'ys1_stress', 'ys2_stress', 'ts_stress', 'elongation']
    
    existing_cols = [col for col in columns if col in data.columns]
    
    if len(existing_cols) < 2:
        print("❌ 상관관계 분석을 위한 충분한 수치형 컬럼이 없습니다.")
        return
    
    # 상관관계 매트릭스
    corr_matrix = data[existing_cols].corr()
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, 
                square=True, fmt='.2f', cbar_kws={'shrink': 0.8})
    plt.title('중경1공장: 주요 변수 상관관계', fontweight='bold', pad=20)
    plt.tight_layout()
    plt.show()
    
    return corr_matrix

print("✅ 분석 함수 정의 완료")
print("\n사용 가능한 함수들:")
print("1. plot_distribution(jg1_data, 'column_name') - 변수 분포 시각화")
print("2. compare_by_quality(jg1_data, 'column_name') - 품질별 변수 비교")
print("3. correlation_analysis(jg1_data) - 상관관계 분석")
```

```python
# 필터링 후 데이터 품질 확인
print("\n📊 필터링 후 데이터 품질 확인")
print("=" * 60)

# 필터링된 컬럼들의 통계 확인
print("필터링된 컬럼들의 기본 통계:")
if actual_columns:
    filtered_stats = jg1_data[actual_columns].describe()
    print(filtered_stats.round(3))
    
    # 각 컬럼의 최솟값이 0보다 큰지 확인
    print(f"\n✅ 0값 제거 검증:")
    for col in actual_columns:
        min_val = jg1_data[col].min()
        zero_count = (jg1_data[col] == 0).sum()
        print(f"  {col}: 최솟값 = {min_val}, 0값 개수 = {zero_count}")

# 품질별 분포 재확인
print(f"\n📈 필터링 후 품질 분포:")
if 'quality' in jg1_data.columns:
    quality_counts_filtered = jg1_data['quality'].value_counts()
    print(f"품질 유형 수: {len(quality_counts_filtered)}개")
    print("상위 5개 품질:")
    for i, (quality, count) in enumerate(quality_counts_filtered.head().items(), 1):
        pct = (count / len(jg1_data)) * 100
        print(f"  {i}. {quality}: {count:,}개 ({pct:.1f}%)")

print(f"\n🎯 최종 중경1공장 분석 데이터: {len(jg1_data):,}개 준비 완료!")
```

# 🎯 중경1공장 X52 계열 세아제강 제품 규격 내 품질별 항복강도 분포 분석

## 분석 목표
- 중경1공장의 X52 계열 제품 중 세아제강 규격을 만족하는 데이터 필터링
- 품질별 항복강도 분포를 Scatterplot으로 시각화
- X52 API 5L 규격 기준 적용


```python
# 🚀 중경1공장 X52 계열 분석을 위한 데이터 로드 및 필터링
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# 한글 폰트 설정
import matplotlib.font_manager as fm
import matplotlib
font_path = "C:/Windows/Fonts/malgun.ttf"
font_prop = fm.FontProperties(fname=font_path)
matplotlib.rcParams['font.family'] = font_prop.get_name()
matplotlib.rcParams['axes.unicode_minus'] = False

# 시각화 스타일 설정
plt.style.use('default')
sns.set_palette("husl")

print("🚀 중경1공장 X52 계열 세아제강 제품 분석 시작")
print("=" * 80)
```

```python
# 🔍 X52 계열 데이터 필터링
print("\n🔍 X52 계열 데이터 필터링 중...")

if quality_col is None:
    print("❌ 품질 컬럼이 없어 분석을 진행할 수 없습니다.")
else:
    # 전체 품질 유형 확인
    print(f"\n📋 전체 품질 유형:")
    quality_counts = jg1_data[quality_col].value_counts()
    for i, (quality, count) in enumerate(quality_counts.items(), 1):
        print(f"{i:2d}. {quality}: {count:,}개")
    
    # X52 계열 데이터 필터링
    x52_mask = jg1_data[quality_col].str.contains('X52', case=False, na=False)
    x52_data = jg1_data[x52_mask].copy()
    
    print(f"\n✅ X52 계열 데이터 필터링 완료:")
    print(f"   전체 데이터: {len(jg1_data):,}개")
    print(f"   X52 계열: {len(x52_data):,}개 ({len(x52_data)/len(jg1_data)*100:.1f}%)")
    
    if len(x52_data) > 0:
        print(f"\n📊 X52 계열 품질 분포:")
        x52_qualities = x52_data[quality_col].value_counts()
        for quality, count in x52_qualities.items():
            print(f"   {quality}: {count:,}개")
```

```python
# 🎯 세아제강 X52 제품 규격 적용
print("\n🎯 세아제강 X52 제품 규격 적용 중...")

# X52 규격 (API 5L 기준)
# 항복강도(YS): 최소 359 MPa (52,000 psi)
# 인장강도(TS): 최소 455 MPa
# 항복강도/인장강도 비율: 최대 0.93

x52_specs = {
    'min_yield_strength': 359,  # MPa
    'min_tensile_strength': 455,  # MPa
    'max_ys_ts_ratio': 0.93
}

print(f"📋 X52 규격 기준:")
print(f"   항복강도(YS): ≥ {x52_specs['min_yield_strength']} MPa")
print(f"   인장강도(TS): ≥ {x52_specs['min_tensile_strength']} MPa")
print(f"   YS/TS 비율: ≤ {x52_specs['max_ys_ts_ratio']}")

# 항복강도 컬럼 찾기
ys_cols = [col for col in x52_data.columns if any(word in col.lower() for word in ['ys', 'yield', 'i_ys'])]
ts_cols = [col for col in x52_data.columns if any(word in col.lower() for word in ['ts', 'tensile', 'i_ts'])]

print(f"\n🔍 발견된 강도 컬럼들:")
print(f"   항복강도 관련: {ys_cols}")
print(f"   인장강도 관련: {ts_cols}")

if not ys_cols:
    print("❌ 항복강도 컬럼을 찾을 수 없습니다.")
    spec_data = x52_data
    ys_col = None
else:
    # 주요 항복강도 컬럼 선택 (i_ys 우선, 없으면 첫 번째)
    ys_col = 'i_ys' if 'i_ys' in ys_cols else ys_cols[0]
    print(f"✅ 항복강도 컬럼으로 '{ys_col}' 사용")
    
    # 규격 내 데이터 필터링
    original_count = len(x52_data)
    
    # 항복강도 기준 적용
    spec_data = x52_data[x52_data[ys_col] >= x52_specs['min_yield_strength']].copy()
    
    # 인장강도 기준 적용 (컬럼이 있는 경우)
    if ts_cols:
        ts_col = 'i_ts' if 'i_ts' in ts_cols else ts_cols[0]
        print(f"✅ 인장강도 컬럼으로 '{ts_col}' 사용")
        spec_data = spec_data[spec_data[ts_col] >= x52_specs['min_tensile_strength']]
        
        # YS/TS 비율 확인
        spec_data['ys_ts_ratio'] = spec_data[ys_col] / spec_data[ts_col]
        spec_data = spec_data[spec_data['ys_ts_ratio'] <= x52_specs['max_ys_ts_ratio']]
    
    print(f"\n✅ 세아제강 규격 적용 완료:")
    print(f"   필터링 전: {original_count:,}개")
    print(f"   규격 내: {len(spec_data):,}개 ({len(spec_data)/original_count*100:.1f}%)")
    
    if len(spec_data) > 0:
        print(f"\n📊 규격 내 품질 분포:")
        spec_qualities = spec_data[quality_col].value_counts()
        for quality, count in spec_qualities.items():
            print(f"   {quality}: {count:,}개")
            
        # 항복강도 통계
        print(f"\n📈 규격 내 항복강도({ys_col}) 통계:")
        print(spec_data[ys_col].describe().round(1))
```

```python
# 🎨 품질별 항복강도 분포 scatterplot 생성
print(f"\n🎨 품질별 항복강도 분포 Scatterplot 생성 중...")

if len(spec_data) == 0 or ys_col is None:
    print("❌ 데이터가 없어 그래프를 생성할 수 없습니다.")
else:
    # 그래프 크기 설정
    plt.figure(figsize=(14, 8))
    
    # 품질별로 다른 색상과 마커 사용
    qualities = spec_data[quality_col].unique()
    colors = plt.cm.Set1(np.linspace(0, 1, len(qualities)))
    markers = ['o', 's', '^', 'D', 'v', '<', '>', 'p', '*', 'h']
    
    for i, quality in enumerate(qualities):
        quality_data = spec_data[spec_data[quality_col] == quality]
        
        # 산점도 그리기
        plt.scatter(
            range(len(quality_data)), 
            quality_data[ys_col],
            c=[colors[i]], 
            marker=markers[i % len(markers)],
            s=60, 
            alpha=0.7, 
            label=f'{quality} (n={len(quality_data)})',
            edgecolors='black',
            linewidth=0.5
        )
        
        # 품질별 평균선 추가
        mean_ys = quality_data[ys_col].mean()
        plt.axhline(y=mean_ys, color=colors[i], linestyle='--', alpha=0.8, linewidth=1.5)
    
    # X52 최소 규격선 추가
    plt.axhline(y=359, color='red', linestyle='-', linewidth=2, alpha=0.8, label='X52 최소 규격 (359 MPa)')
    
    # 그래프 꾸미기
    plt.title('중경1공장: X52 계열 세아제강 제품 규격 내\\n품질별 항복강도 분포 (Scatterplot)', 
              fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('데이터 순서', fontsize=12, fontweight='bold')
    plt.ylabel('항복강도 (MPa)', fontsize=12, fontweight='bold')
    
    # 격자 추가
    plt.grid(True, alpha=0.3)
    
    # 범례 설정
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
    
    # Y축 범위 설정 (최소값보다 약간 아래부터)
    y_min = spec_data[ys_col].min() - 10
    y_max = spec_data[ys_col].max() + 10
    plt.ylim(y_min, y_max)
    
    # 레이아웃 조정
    plt.tight_layout()
    
    # 저장
    filename = '중경1공장_X52계열_세아제강규격내_품질별_항복강도분포_scatterplot.png'
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"💾 그래프를 '{filename}'로 저장했습니다.")
    
    plt.show()
    
    print("✅ 품질별 항복강도 분포 scatterplot 생성 완료!")
```

```python
# 📊 품질별 항복강도 통계 분석
print(f"\n📊 품질별 항복강도 통계 분석:")

if len(spec_data) > 0 and ys_col is not None:
    stats = spec_data.groupby(quality_col)[ys_col].agg(['count', 'mean', 'std', 'min', 'max']).round(1)
    print(stats)
    
    # 상세 분석
    print(f"\n🔍 품질별 상세 분석:")
    for quality in spec_data[quality_col].unique():
        quality_data = spec_data[spec_data[quality_col] == quality][ys_col]
        print(f"\n📊 {quality}:")
        print(f"   - 개수: {len(quality_data)}개")
        print(f"   - 평균: {quality_data.mean():.1f} MPa")
        print(f"   - 표준편차: {quality_data.std():.1f} MPa")
        print(f"   - 범위: {quality_data.min():.1f} ~ {quality_data.max():.1f} MPa")
        print(f"   - 중위수: {quality_data.median():.1f} MPa")
        print(f"   - 25% 분위수: {quality_data.quantile(0.25):.1f} MPa")
        print(f"   - 75% 분위수: {quality_data.quantile(0.75):.1f} MPa")
    
    # 전체 통계
    print(f"\n🎯 전체 규격 내 데이터 통계:")
    print(f"   전체 평균: {spec_data[ys_col].mean():.1f} MPa")
    print(f"   전체 범위: {spec_data[ys_col].min():.1f} ~ {spec_data[ys_col].max():.1f} MPa")
    print(f"   전체 표준편차: {spec_data[ys_col].std():.1f} MPa")
    
    print("\n✅ 중경1공장 X52 계열 세아제강 제품 분석 완료!")
    print("=" * 80)
else:
    print("❌ 분석할 데이터가 없습니다.")
```

# 📊 상위 5개 품질별 두께 분포 분석 (Stripplot)

상위 5개 품질에 대한 두께 분포를 stripplot으로 시각화하고 각 품질별 데이터 개수를 표시합니다.


```python
# 📊 상위 5개 품질별 두께 분포 stripplot 생성
print("📊 상위 5개 품질별 두께 분포 stripplot 생성 중...")

# 현재 데이터에서 상위 5개 품질 선택
if 'jg1_data' in locals() and quality_col is not None:
    # 품질별 개수 확인
    quality_counts = jg1_data[quality_col].value_counts()
    top_5_qualities = quality_counts.head(5)
    
    # 상위 5개 품질 데이터만 필터링
    top5_data = jg1_data[jg1_data[quality_col].isin(top_5_qualities.index)].copy()
    
    print(f"✅ 상위 5개 품질 데이터 필터링 완료: {len(top5_data)}개")
    print(f"\n📋 상위 5개 품질:")
    for i, (quality, count) in enumerate(top_5_qualities.items(), 1):
        pct = (count / len(jg1_data)) * 100
        print(f"  {i}. {quality}: {count:,}개 ({pct:.1f}%)")
    
    # 두께 컬럼 확인
    thickness_col = None
    thickness_candidates = ['p_thick_mm', 'thick', 'thickness']
    for col in thickness_candidates:
        if col in jg1_data.columns:
            thickness_col = col
            break
    
    if thickness_col is None:
        print("❌ 두께 컬럼을 찾을 수 없습니다.")
    else:
        print(f"✅ 두께 컬럼으로 '{thickness_col}' 사용")
        
        # 그래프 생성
        plt.figure(figsize=(14, 8))
        
        # stripplot 생성
        sns.stripplot(data=top5_data, x=quality_col, y=thickness_col, 
                     size=6, alpha=0.7, jitter=True)
        
        # 그래프 제목 및 라벨
        plt.title('중경1공장: 상위 5개 품질별 두께 분포 (Stripplot)', 
                  fontsize=16, fontweight='bold', pad=20)
        plt.xlabel('품질 (Quality)', fontsize=12, fontweight='bold')
        plt.ylabel('두께 (mm)', fontsize=12, fontweight='bold')
        
        # x축 라벨 회전
        plt.xticks(rotation=45, ha='right')
        
        # 격자 추가
        plt.grid(True, alpha=0.3, axis='y')
        
        # 각 품질별 평균선 추가
        for i, quality in enumerate(top_5_qualities.index):
            quality_data = top5_data[top5_data[quality_col] == quality][thickness_col]
            mean_val = quality_data.mean()
            plt.hlines(mean_val, i-0.3, i+0.3, colors='red', linestyles='--', 
                      alpha=0.8, linewidth=2)
        
        # 오른쪽 상단에 품질별 개수 정보 추가
        info_text = "품질별 데이터 개수:\\n"
        for quality, count in top_5_qualities.items():
            info_text += f"{quality}: {count:,}개\\n"
        
        plt.text(0.98, 0.98, info_text, transform=plt.gca().transAxes,
                fontsize=10, verticalalignment='top', horizontalalignment='right',
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
        
        # 평균선 범례 추가
        plt.plot([], [], 'r--', alpha=0.8, linewidth=2, label='평균값')
        plt.legend(loc='upper left')
        
        # 레이아웃 조정
        plt.tight_layout()
        
        # 저장
        filename = '중경1공장_상위5개품질_두께분포_stripplot.png'
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"💾 그래프를 '{filename}'로 저장했습니다.")
        
        plt.show()
        
        print("✅ 상위 5개 품질별 두께 분포 stripplot 생성 완료!")
        
else:
    print("❌ 데이터가 로드되지 않았습니다. 먼저 데이터 로드 셀을 실행해주세요.")
```

```python
# 📈 상위 5개 품질별 두께 통계 분석
print("\n📈 상위 5개 품질별 두께 통계 분석:")

if 'top5_data' in locals() and thickness_col is not None:
    # 품질별 두께 통계
    thickness_stats = top5_data.groupby(quality_col)[thickness_col].agg([
        'count', 'mean', 'std', 'min', 'max', 'median'
    ]).round(2)
    
    print("\n📊 품질별 두께 통계 요약:")
    print(thickness_stats)
    
    # 상세 분석
    print(f"\n🔍 품질별 상세 분석:")
    for quality in top_5_qualities.index:
        quality_data = top5_data[top5_data[quality_col] == quality][thickness_col]
        print(f"\n📊 {quality}:")
        print(f"   - 개수: {len(quality_data)}개")
        print(f"   - 평균: {quality_data.mean():.2f}mm")
        print(f"   - 표준편차: {quality_data.std():.2f}mm")
        print(f"   - 범위: {quality_data.min():.1f} ~ {quality_data.max():.1f}mm")
        print(f"   - 중위수: {quality_data.median():.2f}mm")
        print(f"   - 25% 분위수: {quality_data.quantile(0.25):.2f}mm")
        print(f"   - 75% 분위수: {quality_data.quantile(0.75):.2f}mm")
    
    # 전체 통계
    print(f"\n🎯 전체 상위 5개 품질 두께 통계:")
    print(f"   전체 평균: {top5_data[thickness_col].mean():.2f}mm")
    print(f"   전체 범위: {top5_data[thickness_col].min():.1f} ~ {top5_data[thickness_col].max():.1f}mm")
    print(f"   전체 표준편차: {top5_data[thickness_col].std():.2f}mm")
    
    print("\n✅ 상위 5개 품질별 두께 분포 분석 완료!")
else:
    print("❌ 분석할 데이터가 없습니다.")
```
