```python
# ì¤‘ê²½1ê³µì¥ ì „ìš© ë¶„ì„ í•¨ìˆ˜ë“¤

def plot_distribution(data, column, title=None, bins=30):
    """ìˆ˜ì¹˜í˜• ë³€ìˆ˜ì˜ ë¶„í¬ë¥¼ íˆìŠ¤í† ê·¸ë¨ê³¼ ë°•ìŠ¤í”Œë¡¯ìœ¼ë¡œ ì‹œê°í™”"""
    if column not in data.columns:
        print(f"âŒ '{column}' ì»¬ëŸ¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return
    
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    # íˆìŠ¤í† ê·¸ë¨
    data[column].dropna().hist(bins=bins, ax=axes[0], alpha=0.7, edgecolor='black')
    axes[0].set_title(f'{title or column} ë¶„í¬', fontweight='bold')
    axes[0].set_xlabel(column)
    axes[0].set_ylabel('ë¹ˆë„')
    axes[0].grid(True, alpha=0.3)
    
    # ë°•ìŠ¤í”Œë¡¯
    data[column].dropna().plot(kind='box', ax=axes[1])
    axes[1].set_title(f'{title or column} ë°•ìŠ¤í”Œë¡¯', fontweight='bold')
    axes[1].set_ylabel(column)
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # ê¸°ë³¸ í†µê³„
    print(f"ğŸ“Š {title or column} ê¸°ë³¸ í†µê³„:")
    print(data[column].describe().round(2))

def compare_by_quality(data, column, top_n=5):
    """í’ˆì§ˆë³„ë¡œ íŠ¹ì • ë³€ìˆ˜ë¥¼ ë¹„êµ ë¶„ì„"""
    if 'quality' not in data.columns or column not in data.columns:
        print(f"âŒ í•„ìš”í•œ ì»¬ëŸ¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return
    
    # ìƒìœ„ Nê°œ í’ˆì§ˆ ìœ í˜• ì„ íƒ
    top_qualities = data['quality'].value_counts().head(top_n).index
    filtered_data = data[data['quality'].isin(top_qualities)]
    
    plt.figure(figsize=(12, 6))
    
    # ë°•ìŠ¤í”Œë¡¯
    sns.boxplot(data=filtered_data, x='quality', y=column)
    plt.title(f'ì¤‘ê²½1ê³µì¥: í’ˆì§ˆë³„ {column} ë¹„êµ (ìƒìœ„ {top_n}ê°œ í’ˆì§ˆ)', fontweight='bold')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # í’ˆì§ˆë³„ í†µê³„
    print(f"\nğŸ“Š í’ˆì§ˆë³„ {column} í†µê³„:")
    quality_stats = filtered_data.groupby('quality')[column].agg(['count', 'mean', 'std', 'min', 'max']).round(2)
    print(quality_stats)

def correlation_analysis(data, columns=None):
    """ì£¼ìš” ë³€ìˆ˜ë“¤ ê°„ì˜ ìƒê´€ê´€ê³„ ë¶„ì„"""
    if columns is None:
        columns = ['p_thick_mm', 'ys1_stress', 'ys2_stress', 'ts_stress', 'elongation']
    
    existing_cols = [col for col in columns if col in data.columns]
    
    if len(existing_cols) < 2:
        print("âŒ ìƒê´€ê´€ê³„ ë¶„ì„ì„ ìœ„í•œ ì¶©ë¶„í•œ ìˆ˜ì¹˜í˜• ì»¬ëŸ¼ì´ ì—†ìŠµë‹ˆë‹¤.")
        return
    
    # ìƒê´€ê´€ê³„ ë§¤íŠ¸ë¦­ìŠ¤
    corr_matrix = data[existing_cols].corr()
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, 
                square=True, fmt='.2f', cbar_kws={'shrink': 0.8})
    plt.title('ì¤‘ê²½1ê³µì¥: ì£¼ìš” ë³€ìˆ˜ ìƒê´€ê´€ê³„', fontweight='bold', pad=20)
    plt.tight_layout()
    plt.show()
    
    return corr_matrix

print("âœ… ë¶„ì„ í•¨ìˆ˜ ì •ì˜ ì™„ë£Œ")
print("\nì‚¬ìš© ê°€ëŠ¥í•œ í•¨ìˆ˜ë“¤:")
print("1. plot_distribution(jg1_data, 'column_name') - ë³€ìˆ˜ ë¶„í¬ ì‹œê°í™”")
print("2. compare_by_quality(jg1_data, 'column_name') - í’ˆì§ˆë³„ ë³€ìˆ˜ ë¹„êµ")
print("3. correlation_analysis(jg1_data) - ìƒê´€ê´€ê³„ ë¶„ì„")
```

```python
# ğŸ” X52 ê³„ì—´ ë°ì´í„° í•„í„°ë§
print("\nğŸ” X52 ê³„ì—´ ë°ì´í„° í•„í„°ë§ ì¤‘...")

if quality_col is None:
    print("âŒ í’ˆì§ˆ ì»¬ëŸ¼ì´ ì—†ì–´ ë¶„ì„ì„ ì§„í–‰í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
else:
    # ì „ì²´ í’ˆì§ˆ ìœ í˜• í™•ì¸
    print(f"\nğŸ“‹ ì „ì²´ í’ˆì§ˆ ìœ í˜•:")
    quality_counts = jg1_data[quality_col].value_counts()
    for i, (quality, count) in enumerate(quality_counts.items(), 1):
        print(f"{i:2d}. {quality}: {count:,}ê°œ")
    
    # X52 ê³„ì—´ ë°ì´í„° í•„í„°ë§
    x52_mask = jg1_data[quality_col].str.contains('X52', case=False, na=False)
    x52_data = jg1_data[x52_mask].copy()
    
    print(f"\nâœ… X52 ê³„ì—´ ë°ì´í„° í•„í„°ë§ ì™„ë£Œ:")
    print(f"   ì „ì²´ ë°ì´í„°: {len(jg1_data):,}ê°œ")
    print(f"   X52 ê³„ì—´: {len(x52_data):,}ê°œ ({len(x52_data)/len(jg1_data)*100:.1f}%)")
    
    if len(x52_data) > 0:
        print(f"\nğŸ“Š X52 ê³„ì—´ í’ˆì§ˆ ë¶„í¬:")
        x52_qualities = x52_data[quality_col].value_counts()
        for quality, count in x52_qualities.items():
            print(f"   {quality}: {count:,}ê°œ")
```

```python
# ğŸ¯ ì„¸ì•„ì œê°• X52 ì œí’ˆ ê·œê²© ì ìš©
print("\nğŸ¯ ì„¸ì•„ì œê°• X52 ì œí’ˆ ê·œê²© ì ìš© ì¤‘...")

# X52 ê·œê²© (API 5L ê¸°ì¤€)
# í•­ë³µê°•ë„(YS): ìµœì†Œ 359 MPa (52,000 psi)
# ì¸ì¥ê°•ë„(TS): ìµœì†Œ 455 MPa
# í•­ë³µê°•ë„/ì¸ì¥ê°•ë„ ë¹„ìœ¨: ìµœëŒ€ 0.93

x52_specs = {
    'min_yield_strength': 359,  # MPa
    'min_tensile_strength': 455,  # MPa
    'max_ys_ts_ratio': 0.93
}

print(f"ğŸ“‹ X52 ê·œê²© ê¸°ì¤€:")
print(f"   í•­ë³µê°•ë„(YS): â‰¥ {x52_specs['min_yield_strength']} MPa")
print(f"   ì¸ì¥ê°•ë„(TS): â‰¥ {x52_specs['min_tensile_strength']} MPa")
print(f"   YS/TS ë¹„ìœ¨: â‰¤ {x52_specs['max_ys_ts_ratio']}")

# í•­ë³µê°•ë„ ì»¬ëŸ¼ ì°¾ê¸°
ys_cols = [col for col in x52_data.columns if any(word in col.lower() for word in ['ys', 'yield', 'i_ys'])]
ts_cols = [col for col in x52_data.columns if any(word in col.lower() for word in ['ts', 'tensile', 'i_ts'])]

print(f"\nğŸ” ë°œê²¬ëœ ê°•ë„ ì»¬ëŸ¼ë“¤:")
print(f"   í•­ë³µê°•ë„ ê´€ë ¨: {ys_cols}")
print(f"   ì¸ì¥ê°•ë„ ê´€ë ¨: {ts_cols}")

if not ys_cols:
    print("âŒ í•­ë³µê°•ë„ ì»¬ëŸ¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
    spec_data = x52_data
    ys_col = None
else:
    # ì£¼ìš” í•­ë³µê°•ë„ ì»¬ëŸ¼ ì„ íƒ (i_ys ìš°ì„ , ì—†ìœ¼ë©´ ì²« ë²ˆì§¸)
    ys_col = 'i_ys' if 'i_ys' in ys_cols else ys_cols[0]
    print(f"âœ… í•­ë³µê°•ë„ ì»¬ëŸ¼ìœ¼ë¡œ '{ys_col}' ì‚¬ìš©")
    
    # ê·œê²© ë‚´ ë°ì´í„° í•„í„°ë§
    original_count = len(x52_data)
    
    # í•­ë³µê°•ë„ ê¸°ì¤€ ì ìš©
    spec_data = x52_data[x52_data[ys_col] >= x52_specs['min_yield_strength']].copy()
    
    # ì¸ì¥ê°•ë„ ê¸°ì¤€ ì ìš© (ì»¬ëŸ¼ì´ ìˆëŠ” ê²½ìš°)
    if ts_cols:
        ts_col = 'i_ts' if 'i_ts' in ts_cols else ts_cols[0]
        print(f"âœ… ì¸ì¥ê°•ë„ ì»¬ëŸ¼ìœ¼ë¡œ '{ts_col}' ì‚¬ìš©")
        spec_data = spec_data[spec_data[ts_col] >= x52_specs['min_tensile_strength']]
        
        # YS/TS ë¹„ìœ¨ í™•ì¸
        spec_data['ys_ts_ratio'] = spec_data[ys_col] / spec_data[ts_col]
        spec_data = spec_data[spec_data['ys_ts_ratio'] <= x52_specs['max_ys_ts_ratio']]
    
    print(f"\nâœ… ì„¸ì•„ì œê°• ê·œê²© ì ìš© ì™„ë£Œ:")
    print(f"   í•„í„°ë§ ì „: {original_count:,}ê°œ")
    print(f"   ê·œê²© ë‚´: {len(spec_data):,}ê°œ ({len(spec_data)/original_count*100:.1f}%)")
    
    if len(spec_data) > 0:
        print(f"\nğŸ“Š ê·œê²© ë‚´ í’ˆì§ˆ ë¶„í¬:")
        spec_qualities = spec_data[quality_col].value_counts()
        for quality, count in spec_qualities.items():
            print(f"   {quality}: {count:,}ê°œ")
            
        # í•­ë³µê°•ë„ í†µê³„
        print(f"\nğŸ“ˆ ê·œê²© ë‚´ í•­ë³µê°•ë„({ys_col}) í†µê³„:")
        print(spec_data[ys_col].describe().round(1))
```

# ğŸ“Š ìƒìœ„ 5ê°œ í’ˆì§ˆë³„ ë‘ê»˜ ë¶„í¬ ë¶„ì„ (Stripplot)

ìƒìœ„ 5ê°œ í’ˆì§ˆì— ëŒ€í•œ ë‘ê»˜ ë¶„í¬ë¥¼ stripplotìœ¼ë¡œ ì‹œê°í™”í•˜ê³  ê° í’ˆì§ˆë³„ ë°ì´í„° ê°œìˆ˜ë¥¼ í‘œì‹œí•©ë‹ˆë‹¤.


```python
# ğŸ“Š ìƒìœ„ 5ê°œ í’ˆì§ˆë³„ ë‘ê»˜ ë¶„í¬ stripplot ìƒì„±
print("ğŸ“Š ìƒìœ„ 5ê°œ í’ˆì§ˆë³„ ë‘ê»˜ ë¶„í¬ stripplot ìƒì„± ì¤‘...")

# í˜„ì¬ ë°ì´í„°ì—ì„œ ìƒìœ„ 5ê°œ í’ˆì§ˆ ì„ íƒ
if 'jg1_data' in locals() and quality_col is not None:
    # í’ˆì§ˆë³„ ê°œìˆ˜ í™•ì¸
    quality_counts = jg1_data[quality_col].value_counts()
    top_5_qualities = quality_counts.head(5)
    
    # ìƒìœ„ 5ê°œ í’ˆì§ˆ ë°ì´í„°ë§Œ í•„í„°ë§
    top5_data = jg1_data[jg1_data[quality_col].isin(top_5_qualities.index)].copy()
    
    print(f"âœ… ìƒìœ„ 5ê°œ í’ˆì§ˆ ë°ì´í„° í•„í„°ë§ ì™„ë£Œ: {len(top5_data)}ê°œ")
    print(f"\nğŸ“‹ ìƒìœ„ 5ê°œ í’ˆì§ˆ:")
    for i, (quality, count) in enumerate(top_5_qualities.items(), 1):
        pct = (count / len(jg1_data)) * 100
        print(f"  {i}. {quality}: {count:,}ê°œ ({pct:.1f}%)")
    
    # ë‘ê»˜ ì»¬ëŸ¼ í™•ì¸
    thickness_col = None
    thickness_candidates = ['p_thick_mm', 'thick', 'thickness']
    for col in thickness_candidates:
        if col in jg1_data.columns:
            thickness_col = col
            break
    
    if thickness_col is None:
        print("âŒ ë‘ê»˜ ì»¬ëŸ¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
    else:
        print(f"âœ… ë‘ê»˜ ì»¬ëŸ¼ìœ¼ë¡œ '{thickness_col}' ì‚¬ìš©")
        
        # ê·¸ë˜í”„ ìƒì„±
        plt.figure(figsize=(14, 8))
        
        # stripplot ìƒì„±
        sns.stripplot(data=top5_data, x=quality_col, y=thickness_col, 
                     size=6, alpha=0.7, jitter=True)
        
        # ê·¸ë˜í”„ ì œëª© ë° ë¼ë²¨
        plt.title('ì¤‘ê²½1ê³µì¥: ìƒìœ„ 5ê°œ í’ˆì§ˆë³„ ë‘ê»˜ ë¶„í¬ (Stripplot)', 
                  fontsize=16, fontweight='bold', pad=20)
        plt.xlabel('í’ˆì§ˆ (Quality)', fontsize=12, fontweight='bold')
        plt.ylabel('ë‘ê»˜ (mm)', fontsize=12, fontweight='bold')
        
        # xì¶• ë¼ë²¨ íšŒì „
        plt.xticks(rotation=45, ha='right')
        
        # ê²©ì ì¶”ê°€
        plt.grid(True, alpha=0.3, axis='y')
        
        # ê° í’ˆì§ˆë³„ í‰ê· ì„  ì¶”ê°€
        for i, quality in enumerate(top_5_qualities.index):
            quality_data = top5_data[top5_data[quality_col] == quality][thickness_col]
            mean_val = quality_data.mean()
            plt.hlines(mean_val, i-0.3, i+0.3, colors='red', linestyles='--', 
                      alpha=0.8, linewidth=2)
        
        # ì˜¤ë¥¸ìª½ ìƒë‹¨ì— í’ˆì§ˆë³„ ê°œìˆ˜ ì •ë³´ ì¶”ê°€
        info_text = "í’ˆì§ˆë³„ ë°ì´í„° ê°œìˆ˜:\\n"
        for quality, count in top_5_qualities.items():
            info_text += f"{quality}: {count:,}ê°œ\\n"
        
        plt.text(0.98, 0.98, info_text, transform=plt.gca().transAxes,
                fontsize=10, verticalalignment='top', horizontalalignment='right',
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
        
        # í‰ê· ì„  ë²”ë¡€ ì¶”ê°€
        plt.plot([], [], 'r--', alpha=0.8, linewidth=2, label='í‰ê· ê°’')
        plt.legend(loc='upper left')
        
        # ë ˆì´ì•„ì›ƒ ì¡°ì •
        plt.tight_layout()
        
        # ì €ì¥
        filename = 'ì¤‘ê²½1ê³µì¥_ìƒìœ„5ê°œí’ˆì§ˆ_ë‘ê»˜ë¶„í¬_stripplot.png'
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"ğŸ’¾ ê·¸ë˜í”„ë¥¼ '{filename}'ë¡œ ì €ì¥í–ˆìŠµë‹ˆë‹¤.")
        
        plt.show()
        
        print("âœ… ìƒìœ„ 5ê°œ í’ˆì§ˆë³„ ë‘ê»˜ ë¶„í¬ stripplot ìƒì„± ì™„ë£Œ!")
        
else:
    print("âŒ ë°ì´í„°ê°€ ë¡œë“œë˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤. ë¨¼ì € ë°ì´í„° ë¡œë“œ ì…€ì„ ì‹¤í–‰í•´ì£¼ì„¸ìš”.")
```

```python
# ğŸ“ˆ ìƒìœ„ 5ê°œ í’ˆì§ˆë³„ ë‘ê»˜ í†µê³„ ë¶„ì„
print("\nğŸ“ˆ ìƒìœ„ 5ê°œ í’ˆì§ˆë³„ ë‘ê»˜ í†µê³„ ë¶„ì„:")

if 'top5_data' in locals() and thickness_col is not None:
    # í’ˆì§ˆë³„ ë‘ê»˜ í†µê³„
    thickness_stats = top5_data.groupby(quality_col)[thickness_col].agg([
        'count', 'mean', 'std', 'min', 'max', 'median'
    ]).round(2)
    
    print("\nğŸ“Š í’ˆì§ˆë³„ ë‘ê»˜ í†µê³„ ìš”ì•½:")
    print(thickness_stats)
    
    # ìƒì„¸ ë¶„ì„
    print(f"\nğŸ” í’ˆì§ˆë³„ ìƒì„¸ ë¶„ì„:")
    for quality in top_5_qualities.index:
        quality_data = top5_data[top5_data[quality_col] == quality][thickness_col]
        print(f"\nğŸ“Š {quality}:")
        print(f"   - ê°œìˆ˜: {len(quality_data)}ê°œ")
        print(f"   - í‰ê· : {quality_data.mean():.2f}mm")
        print(f"   - í‘œì¤€í¸ì°¨: {quality_data.std():.2f}mm")
        print(f"   - ë²”ìœ„: {quality_data.min():.1f} ~ {quality_data.max():.1f}mm")
        print(f"   - ì¤‘ìœ„ìˆ˜: {quality_data.median():.2f}mm")
        print(f"   - 25% ë¶„ìœ„ìˆ˜: {quality_data.quantile(0.25):.2f}mm")
        print(f"   - 75% ë¶„ìœ„ìˆ˜: {quality_data.quantile(0.75):.2f}mm")
    
    # ì „ì²´ í†µê³„
    print(f"\nğŸ¯ ì „ì²´ ìƒìœ„ 5ê°œ í’ˆì§ˆ ë‘ê»˜ í†µê³„:")
    print(f"   ì „ì²´ í‰ê· : {top5_data[thickness_col].mean():.2f}mm")
    print(f"   ì „ì²´ ë²”ìœ„: {top5_data[thickness_col].min():.1f} ~ {top5_data[thickness_col].max():.1f}mm")
    print(f"   ì „ì²´ í‘œì¤€í¸ì°¨: {top5_data[thickness_col].std():.2f}mm")
    
    print("\nâœ… ìƒìœ„ 5ê°œ í’ˆì§ˆë³„ ë‘ê»˜ ë¶„í¬ ë¶„ì„ ì™„ë£Œ!")
else:
    print("âŒ ë¶„ì„í•  ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
```
